### Nosso exemplo:

- Faça o download do nosso PDF

- Local arquivo: http://www.vale.com/PT/investors/information-market/quarterly-results/ResultadosTrimestrais/Vale_IFRS_1T21_BRL_v26042021_vf.pdf

```python
import tabula
import pandas as pd
```

```python
lista_tabelas = tabula.read_pdf("ResultadoVale.pdf", pages = 'all') 

#pages = '12' para ver apenas na pagina 12 do pdf. Mais rapido e eficiente

print(f'Existem {len(lista_tabelas)} tabelas nesse PDF. (teoricamente)')

```

```python
for tabela in lista_tabelas:
    display(tabela)
```

### Mas nem sempre a tabela vai vir bonitinha

```python
lista_tabelas2 = tabula.read_pdf("ResultadoVale.pdf", pages = '10')

print(f'Achei apenas {len(lista_tabelas2)} tabela nessa página')
```

```python
tabela = lista_tabelas2[0]
# display(tabela)

#Veio duas tabelas falando que é uma so. Da uma reparada no arquivo fonte.
```

```python
tabela = lista_tabelas2[0]
tabela.columns = tabela.iloc[0]

tabela[[0,1]] = tabela['Variação percentual'].str.split(' ', expand = True) # Arrumando a coluna de percentual

tabela = tabela[1:9] # Separando as duas tabelas

tabela = tabela.set_index('R$ milhões')  # Colocando R$ milhões no titulo

tabela.columns = tabela.iloc[0]
tabela = tabela[1:] #Ajuste final

tabela = tabela.drop('1T21/4T20 1T21/1T20', axis = 1)

display(tabela)
```

### E se ele não tiver reconhecendo a tabela?

2 parâmetros para testar:

- lattice
- guess

```python
# lattice
lista_df = tabula.read_pdf("ResultadoVale.pdf", pages="10", lattice=True)
display(lista_df[0])

print("*" * 50)

# guess
lista_df = tabula.read_pdf("ResultadoVale.pdf", pages="10", guess=False)
display(lista_df[0])
```
